```
import pandas as pd

campaign = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv')

campaign['contact'].unique()
```

I can look at the data and tell that some days of the week or some months produce better results than others.

I'm wondering if it's possible for us to see if those results are true for all customers, or if some types of customers respond better on certain days than others?

Should we only call single people on Saturdays? Does it make sense to call students at all?

One other thing we should probably address, does contacting people too frequently for these marketing campaigns have an adverse affect on the outcome?

```
import plotly.express as px

marital_day_counts = campaign.groupby(['day_of_week', 'marital']).size().reset_index(name='count')

fig = px.bar(
    marital_day_counts,
    x='day_of_week',
    y='count',
    color='marital',
    title='Marital Status by Day of the Week',
    labels={'day_of_week': 'Day of the Week', 'count': 'Count', 'marital': 'Marital Status'},
    barmode='group'
)

fig.show()

marital_poutcome_counts = campaign.groupby(['poutcome', 'marital']).size().reset_index(name='count')

fig1 = px.bar(
    marital_poutcome_counts,
    x='poutcome',
    y='count',
    color='marital',
    title='Marital Status by Outcome of Previous Campaign',
    labels={'poutcome': 'Outcome of the Previous Marketing Campaign', 'count': 'Count', 'marital': 'Marital Status'},
    barmode='group'
)

fig1.show()
```

```
job_y_counts = campaign.groupby(['y', 'job']).size().reset_index(name='count')

fig2 = px.bar(
    job_y_counts,
    x='y',
    y='count',
    color='job',
    title='What Kind of Jobs have subscribed to a term deposit',
    labels={'y': 'Have they Subscribed to a term deposit', 'count': 'Count', 'job': 'Job'},
    barmode='group'
)

fig2.show()

job_poutcome_counts = campaign.groupby(['poutcome', 'job']).size().reset_index(name='count')

fig3 = px.bar(
    job_poutcome_counts,
    x='poutcome',
    y='count',
    color='job',
    title='How Did Jobs Respond to the Previous Marketing Campaign',
    labels={'poutcome': 'Response to Previous Marketing Campaing', 'count': 'Count', 'job': 'Job'},
    barmode='group'
)

fig3.show()
```

```
# VISUAL THAT SHOWS SUCESS RATIO BASED ON MARITAL STATUS

marital_success_failure = campaign[campaign['poutcome'] != 'nonexistent'].groupby(['marital', 'poutcome']).size().unstack(fill_value=0)


marital_success_failure['success_to_failure_ratio'] = marital_success_failure['success'] / marital_success_failure['failure']


marital_success_failure = marital_success_failure.reset_index()


fig2 = px.bar(
    marital_success_failure,
    x='marital',
    y='success_to_failure_ratio',
    color = 'marital',
    title='Success-to-Failure Ratio by Marital Status',
    labels={'success_to_failure_ratio': 'Success per 1 Failure', 'marital': 'Marital Status'},
    text_auto=True  # Show ratio values on the bars
)

fig2.show()
```

```
import pandas as pd
import plotly.express as px
url = "https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv"
df = pd.read_csv(url, delimiter=";")

# Convert categorical columns to lowercase for consistency
df.columns = df.columns.str.lower()

# Convert "month" and "day_of_week" to categorical types for proper ordering
month_order = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
day_order = ["mon", "tue", "wed", "thu", "fri"]  # Only these appear in the dataset

#df["month"] = pd.Categorical(df["month"], categories=month_order, ordered=True)
#df["day_of_week"] = pd.Categorical(df["day_of_week"], categories=day_order, ordered=True)

# Create a new column for response (0 for "no", 1 for "yes")
df["response"] = df["poutcome"].map({"failure": 0, "success": 1})

# Define customer segments to analyze
segments = ["marital", "education", "job"]

# Create facet plots for response rate by day of the week for different segments
facet_plots = {}
for segment in segments:
    fig = px.bar(
        df.groupby(["day_of_week", segment])["response"].mean().reset_index(),
        x="day_of_week",
        y="response",
        color=segment,
        barmode="group",
        facet_col=segment,
        title=f"Response Rate by Day of Week and {segment.capitalize()}",
        labels={"response": "Response Rate", "day_of_week": "Day of the Week"},
    )
    facet_plots[segment] = fig

# Display plots
facet_plots
```
